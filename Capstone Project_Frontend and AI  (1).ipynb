```python
%%writefile app.py
import streamlit as st
import pandas as pd
import joblib

st.title("Personal Finance Coach")

# 1. File Uploader Component
uploaded = st.file_uploader("Upload your bank statement (.csv)", type="csv")

# 2. Financial Goal Checkboxes & Input Fields
goals = st.multiselect("Select your goal(s)",
    ["Pay off debt", "Save for a house", "Pay tuition", "Build emergency fund"])
custom_goal = st.text_input("Or type a custom goal")
target = st.number_input("Target savings amount ($)", min_value=0)
deadline = st.date_input("Goal deadline")

if uploaded is not None:
    df = pd.read_csv(uploaded)
    st.write("### Raw Uploaded Data Preview:")
    st.dataframe(df.head())
```

```python
import os
from google import genai

# 1. Inject API key into environment
os.environ["GEMINI_API_KEY"] = "Your_API_Key"

# 2. Initialize the client 
client = genai.Client()

print("Gemini client successfully initialized!")
```

```python
# 1. Test data
test_savings_goal = "Save for Tuition fee"
test_target_amount = 5000
test_months_left = 10
test_reducible_spend_summary = """
- Food & Dining: $200.00
- Shopping: $200.00
- Entertainment: $75.00
- Transportation: $130.00
"""

# 2. Build the structured prompt template
finance_prompt = f"""
You are an expert personal finance coach. Analyze the user's financial situation and provide highly actionable, specific advice to help them meet their goal.

Financial Goal: {test_savings_goal}
Target Savings Amount: ${test_target_amount}
Time Frame: {test_months_left} months
Required Monthly Savings Rate: ${round(test_target_amount / test_months_left, 2)}/month

Here is a summary of their current monthly flexible (reducible) spending:
{test_reducible_spend_summary}

Provide a structured response containing:
1. An assessment of whether their goal is realistic based on their reducible spend.
2. 3 highly specific, creative strategies to cut spending in their highest discretionary categories without touching essentials (like Rent, Utilities, or Healthcare).
3. A clear, motivational wrap-up sentence.
"""

# 3. Request completion from the Gemini 2.5 Flash model
print("Sending data to Gemini... Please wait...\n")
response = client.models.generate_content(
    model='gemini-2.5-flash',
    contents=finance_prompt,
)

# 4. Print out your financial suggestions
print("--- AI Finance Coach Response ---")
print(response.text)
```
